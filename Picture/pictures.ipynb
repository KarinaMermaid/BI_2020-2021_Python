```python
from PIL import Image, ImageFilter
import numpy as np
```

```python
pict = Image.open(path)
pict
```

```python
data = np.array(pict)
```

```python
def convolve (data, kernel):
    modified = np.zeros_like(data)
    for row in range(1, data.shape[0] - 2):
        for col in range(1, data.shape[1] - 2):
            modified[row, col] = (data[row:row + 3, col:col + 3] * kernel).mean()
    return modified
kernel = np.ones((3,3))
plt.imshow(convolve(data, kernel))
plt.axis('off')
```
